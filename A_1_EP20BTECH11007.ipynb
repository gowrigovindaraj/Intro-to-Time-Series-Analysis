We see that the time series progresses as approximately alternatively increasing and decreasing. $\therefore$, we can assume that the ACF coefficient is close to 0

```python
plt.figure(figsize=(10,8))
plt.plot(df[0:12],df[1:13], 'o')
plt.xlabel('X(t)')
plt.ylabel('X(t+1)')
plt.title('Plot of X(t) vs X(t+1)')
plt.show()
```

Similar to the previous question, we can assume that, looking at the plot, there is no linear trend.

```python
def autocorrelation_function(data, lag):
    mean = np.mean(data)
    gamma_h = 0
    gamma_0 = 0
    for i in range(lag, len(data)):
        gamma_h += (data[i]-mean)*(data[i-lag]-mean)
    for i in range(0, len(data)):
        gamma_0 += (data[i]-mean)**2
    return gamma_h/gamma_0
```

```python
acf = autocorrelation_function(df, 1)
```

```python
df = pd.read_csv('AirlinePassengers.csv')
```

```python
df
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['#Passengers'], '-o')
plt.xlabel('Time')
plt.ylabel('Passengers')
plt.title('Airline Passengers')
plt.show()
```

```python
acf = []
for i in range(0,75):
    acf.append(autocorrelation_function(df['#Passengers'], i))
```

```python
plt.figure(figsize=(10,8))
plt.stem(acf, '-o')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function for Airline Passengers Dataset')
plt.show()
```

There is a cyclical trend in the Autocorrelation function through almost every year. We see the value go down as the lag increases.

#### equation

```python
x = df.to_numpy()
```
